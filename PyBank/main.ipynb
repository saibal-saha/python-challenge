```python
#Importing the necesssary libraries

from pathlib import Path
import csv



#defining the data path for the csv file and reading the csv file
file_path = Path('budget_data.csv')

# definiing empty lists to hold the value of profit & loss and dates data
profit_loss = []
dates = []

# Oening the required file defined by the file_path as a read only file and specifying that the data inside the file is separated by comma

with open(file_path, 'r') as file:
    
    csvreader = csv.reader(file, delimiter=',')
    header = next(csvreader)
    
    
    for row in csvreader:
#        print(row)
# creating a variable profit and storing the profit and loss values as integer
        profit= int(row[1])
        date= str(row[0])
# iterate through each row and populate the two list profit_loss and dates with the profit and loss data respectively
        profit_loss.append(profit)
        dates.append(date)
#        print(date)

#Initializing the variables for the required calculations 
        
total_profit = 0
total_months = 0
average_change = 0
total_change = 0
max_increase_in_profit = 0
max_decrease_in_profit = 0
    
#iterating throught the list profit_loss to get the total number of months and total profit
for profit in profit_loss:
    total_months += 1
    total_profit += profit
    
#print(type(profit_loss))

#initializing the x value as 0
x = 0

#creating an empty list to store the changes in profit and loss data
profit_change_list= []


# using the while statement calculating the change in profit by iterating through each row
while x < (len(profit_loss) -1):
    change = profit_loss[x+1] - profit_loss[x]
# adding the changes in profit and loss to the previously defined empty list
    profit_change_list.append(change)
    x += 1

#    print(f' Yearly Change = {change}')
#calculating the total and the average change of the pro and loss data
    total_change += change
    average_change = total_change/(len(profit_loss)-1)


#calculating the maximum and minimum changes in profit 
   
for profit_change in profit_change_list:
    if max_decrease_in_profit == 0:
        max_decrease_in_profit = profit_change
    elif profit_change < max_decrease_in_profit:
        max_decrease_in_profit = profit_change
    elif profit_change > max_increase_in_profit:
        max_increase_in_profit = profit_change

        
#Inserting a value 0 at the top of the list of change in profit and loss in order to make the list equal to the list of dates
profit_change_list.insert(0, 0)   

#print(profit_change_list)
        
        
#initializing a dictionary by setting the profit_change_list as the key and the list of dates as value
dict_profit_loss_change = {profit_change_list[i]: dates[i] for i in range(len(profit_change_list))}
#print(f'dict values  = {dict_profit_loss_change}')


#getting the date value by setting the key as the maximum decrease in profits
value = dict_profit_loss_change[(max_decrease_in_profit)]
#print(value)

#getting the date value by setting the key as the maximum increase in profits
value1 = dict_profit_loss_change[(max_increase_in_profit)]
#print(value1)


# Set output file path and file name. this is where the output data will be written
output_file_path = Path("budget_data_output.txt")



# printing the required output
print('Financial Analysis')
print()
print('----------------------------')
print()
print(f'Total Months: {len(profit_loss)}')
print()
print(f'Total: ${total_profit}')
print()
print(f'Average Change: ${average_change:.2f}')
print()
print(f'Greatest Increase in Profits: {value1} (${max_increase_in_profit})')
print()
print(f'Greatest Decrease in Profits: {value} (${max_decrease_in_profit})')

#Open the output path as a file object 
# opening the budget_data_output.txt file in write mode and adding the necessary results, 
with open(output_file_path, 'w') as file:
    file.write('Financial Analysis\n')
    file.write('\n')
    file.write('----------------------------\n')
    file.write('\n')
    file.write(f'Total Months: {len(profit_loss)}\n')
    file.write('\n')
    file.write(f'Total: ${total_profit}\n')
    file.write('\n')
    file.write(f'Average Change: ${average_change:.2f}\n')
    file.write('\n')
    file.write(f'Greatest Increase in Profits: {value1} (${max_increase_in_profit})\n')
    file.write('\n')
    file.write(f'Greatest Decrease in Profits: {value} (${max_decrease_in_profit})')
    
    
        
```
